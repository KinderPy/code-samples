# Problem 1.2

This problem asks us to plot and analyze two functions.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(2,7,1001)
f1 = np.exp(x)
f2 = x**3.5

plt.plot(x,f1,'r-', lw=2, label=r'$e^x$')
plt.plot(x,f2, 'b--', lw=2, label=r'$x^{3.5}$')
plt.title('Similar Functions', weight='bold')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
plt.semilogy(x,f1,'r-', lw=2, label=r'$e^x$')
plt.semilogy(x,f2, 'b--', lw=2, label=r'$x^{3.5}$')
plt.title('Similar Functions?', weight='bold')
plt.xlabel('x')
plt.ylabel('log(y)')
plt.legend()
```

```python
plt.loglog(x,f1,'r-', lw=2, label=r'$e^x$')
plt.loglog(x,f2, 'b--', lw=2, label=r'$x^{3.5}$')
plt.title('Similar Functions?', weight='bold')
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.legend()
```

# Summary

Plotting the functions $e^x$ and $x^{3.5}$ on linear axes does not reveal much difference.  However, in the semilog plot, $e^x$ is linear while $x^{3.5}$ is curved.  In the log-log plot, $x^{3.5}$ is linear, while $e^x$ is curved.  Semilog and log-log plots allow us to distinguish these two types of functions: "exponential" and "power law".
